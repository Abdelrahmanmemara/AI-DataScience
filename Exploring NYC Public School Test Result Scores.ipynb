Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

# The questions that we have to answer

1) Which NYC schools have the best math results?

2) What are the top 10 performing schools based on the combined SAT scores?

3) Which single borough has the largest standard deviation in the combined SAT score?

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
schools.info()
```

```python
schools.describe()
```

# First Question:- Which NYC schools have the best math results?


```python
more_than_80_percent = 800 * 0.80

best_math_schools = schools.loc[:, ['school_name', 'average_math']]

best_math_schools = best_math_schools[best_math_schools['average_math'] >= more_than_80_percent]

best_math_schools = best_math_schools.sort_values('average_math', ascending = False)

best_math_schools.head(50)
```

- The best performing school in the math section of the SAT in the whole city of New York is **Stuyvesant High School**.

- Only 10 schools in New York averaged more than 80% in the math section  of ther SAT.

# Second Question:- What are the top 10 performing schools based on the combined SAT scores?

```python
schools['total_SAT'] = round((schools['average_math'] + schools['average_reading'] + schools['average_writing'])/ 3, 1)

schools.head()
```

```python
top_10_schools = schools.loc[:, ['school_name', 'total_SAT']].sort_values('total_SAT', ascending = False)

top_10_schools.head(10)
```

- Above is the best performing schools overall in the SATs. 
- Again, and by quite a difference **Stuyvesant High School** is the best performing school in New York.


# Third Question:- Which single borough has the largest standard deviation in the combined SAT score?

```python
information_about_boroughs = schools.groupby('borough')['total_SAT'].agg(['count', 'mean','std'])

largest_std_dev = information_about_boroughs[information_about_boroughs['std'] == information_about_boroughs['std'].max()]

columns = ['num_schools', 'average_SAT', 'std_SAT']

largest_std_dev = largest_std_dev.rename(columns= {'count': 'num_schools', 'mean': 'average_SAT', 'std': 'std_SAT'})

print(largest_std_dev)
```

**The borough with the largest standard deviation in combined SAT scores is Manhattan**
